```python
# Import libraries 
import matplotlib.pyplot as plt #used for plotting data 
import numpy as np #used for mathematical operations
import pandas as pd #used to loading CSV data
```

```python
load_df = pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/fundamentals/football_players.csv', index_col='Name')
df = pd.DataFrame(load_df)
print(df.head)
```

```python
columns_to_convert = [ 'Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control', 'Composure','Curve','Dribbling', 'Finishing', 'Free kick accuracy', 'Heading accuracy', 'Jumping', 'Long shots', 'Penalties', 'Shot power','Sliding tackle','Sprint speed', 'Stamina','Standing tackle', 'Strength', 'Positioning', 'Reactions', 'Vision', 'Volleys']
#for col in columns_to_convert:
    #if col == str():
        #df[col] = df[col].str.split('+').str[0].astype('Int64')
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce').astype('Int64')
#Extract and convert the numbers
#df[columns_to_convert] = df[columns_to_convert].str.split('+').str[0].astype('Int64')
for col in columns_to_convert:
    df[col] = df[col].fillna(0).astype('Int64')
```

```python
#Inspect for Correlation between Columns.
correlation = df["Acceleration"].corr(df["Sprint speed"])
print(correlation)
```

```python
df['Technique'] = (((((df['Dribbling']*1.9) + (df['Agility']*1.9) + (df['Balance']*1.9) + (df['Ball control']*1.9) + (df['Composure']*1.9))/190)*100) + ((((df['Positioning']*1.8) +  (df['Reactions']*1.8))/180)*100) + ((((df['Vision']*1.6))/160)*100))//8
df['Clinical finishing'] = ((df['Curve']*1.5) + (df['Finishing']*1.9) + (df['Free kick accuracy']*1.75) + (df['Heading accuracy']*1.9) + (df['Long shots']*1.7) +  (df['Penalties']*1.85) + (df['Shot power']*1.8) + (df['Positioning']*1.9) + (df['Volleys']*1.8))/9
df['Physicality'] =  ((df['Aggression']*1.9) + (df['Jumping']*1.5) + (df['Sliding tackle']*1.7) + (df['Standing tackle']*1.7)+ (df['Stamina']*1.9) + (df['Strength']*1.9))/6
df['Pace'] = ((df['Acceleration']*1.9) + (df['Sprint speed']*1.9))/2

print(df['Technique'].head())
print(df['Clinical finishing'].head())
print(df['Physicality'].head())
print(df['Pace'].head())
```

```python
Techmean = df['Technique'].mean()
Techstd = df['Technique'].std()
TechQ1 = df['Technique'].quantile(0.25)
TechQ3 = df['Technique'].quantile(0.75)
TechIQR = TechQ3 - TechQ1 

print(Techmean)
print(Techstd)
print(TechQ1)
print(TechQ3)
print(TechIQR)
```

```python
Cfmean = df['Clinical finishing'].mean()
Cfstd = df['Clinical finishing'].std()
CfQ1 = df['Clinical finishing'].quantile(0.25)
CfQ3 = df['Clinical finishing'].quantile(0.75)
CfIQR = CfQ3 - CfQ1 

print(Cfmean)
print(Cfstd)
print(CfQ1)
print(CfQ3)
print(CfIQR)
```

```python
Phymean = df['Physicality'].mean()
Phystd = df['Physicality'].std()
PhyQ1 = df['Physicality'].quantile(0.25)
PhyQ3 = df['Physicality'].quantile(0.75)
PhyIQR = PhyQ3 - PhyQ1 

print(Phymean)
print(Phystd)
print(PhyQ1)
print(PhyQ3)
print(PhyIQR)
```

```python
AccQ1 = df['Acceleration'].quantile(0.25)
AccQ3 = df['Acceleration'].quantile(0.75)
AccIQR = AccQ3 - AccQ1
df['Acceleration'].describe()
```

```python
PaceQ1 = df['Pace'].quantile(0.25)
PaceQ3 = df['Pace'].quantile(0.75)
PaceIQR = PaceQ3 - PaceQ1
df['Pace'].describe()
```

```python
lower_bound = TechQ1 - 30 * TechIQR
upper_bound = TechQ3 + 30 * TechIQR
lower_bound1 = CfQ1 - 30 * CfIQR
upper_bound1 = CfQ3 + 30 * CfIQR
lower_bound2 = AccQ1 - 50 * AccIQR
upper_bound2 = AccQ3 + 30 * AccIQR
lower_bound3 =PhyQ1 - 50 * PhyIQR
upper_bound3 = PhyQ3 + 30 * PhyIQR
lower_bound4 = PaceQ1 - 30 * PaceIQR
upper_bound4 = PaceQ3 + 30 * PaceIQR

df_clean = df[(df['Technique'] >= lower_bound) & (df['Technique'] <= upper_bound) & (df['Clinical finishing'] >= lower_bound1) & (df['Clinical finishing'] <= upper_bound1) & (df['Acceleration'] >= lower_bound2) & (df['Acceleration'] <= upper_bound2) & (df['Physicality'] >= lower_bound3) & (df['Physicality'] <= upper_bound3) & (df['Pace'] >= lower_bound4) & (df['Pace'] <= upper_bound4)]
df_clean.shape
```
